## 1. Introduction and hypothesis

Our project takes heavy inspiration from Bamman et al's ["Learning Latent Personas of Film Characters"](https://aclanthology.org/P13-1035.pdf). Their work attempts to examine narrative stories through a character lens, rather than the plot lens that previous computational work has been oriented around. Their research focuses on using latent variable models to learn character personas, based on the action verbs, passive verbs, and attributes associated with a character in movie summaries.

While Bamman et al. focused on movie summaries, our project takes a look at a literature dataset instead. Our project aims to answer the following research questions: can we accurately classify a novel’s genre based on its plot summary? Can we also accurately classify a novel’s genre through the main characters’ description, via its associated action verbs (the verbs associated with what a character is doing), patient verbs (verbs that describe actions that are being done to the characters) and adjectives (any adjectives that describe the character). How do the two classification problems compare in terms of accuracy and precision of results? These research questions are incredibly interesting as they analyze fundamental aspects of any text, the plot and the characters, to determine which is more impactful in defining a novel’s genre.

Our research questions are also important because we are using plot and character information gathered from plot summaries, a more limited text source, rather than entire novels in order to predict genre. We believe such a classification model would be helpful for literary scholars to glean broad information about new texts without having to read the entirety of them which can be an arduous process. At the same time, the benefits of such a model would expand beyond academia. Book realtors and librarians may gain a new categorization method for their book catalogs and develop stronger recommendations for genre based on shared plot and/or character components from plot summaries. Books sometimes fall victim to [genre label misclassifications](https://www.writersdigest.com/write-better-fiction/writing-mistakes-writers-make-inaccurate-genre-labels), resulting in poor book recommendations and decreased reader satisfaction. Such misclassifications hinder the reading experience as readers may have misconceptions about the novel and incorrectly assume it aligns with their reading preferences when in fact it does not due to no fault of their own. By creating a prediction model that can identify genre from plot summaries, this may help reduce this problem. This improvement means that the results could be of interest to online review and recommendation systems such as Goodreads, which are widely utilized by contemporary readers who would appreciate more accurate genre labels.

We hypothesize that plot would be a better predictor of genre than character. Not to diminish the significance of the characters in books, we think that the order of events and specific plot points or tropes as described within the plot summaries are important characteristics of a genre. For example, when it comes to the romance genre, happily ever afters come to mind, which are considered more prominent in that genre in comparison to others, at least according to popular culture. Following that logic, we believe that plot is crucial to define what makes up a genre.

## 2. Data and methods

Our data consists of book summary information, pulled from the following site: https://www.cs.cmu.edu/~dbamman/booksummaries.html. Bamman and Smith developed this dataset for their 2013 research article, "New Alignment Methods for Discriminative Book Summarization.” The original dataset includes the following fields, all of which are of type string: Wikipedia ID, Freebase ID, Book title, Book author, Publication date, Genres, and Plot Summary. While the plot summary descriptions were collected from Wikipedia, the metadata, which includes the book title, author, publication date, and genres, was gathered from Freebase. The database originally consisted of 16,559 books, but we cleaned up the data to remove any entries that had missing values. Then we chose to take a random sample of the remaining entries in order to condense the size of the data, making it more manageable to conduct our analysis on it.

Despite the inclusion of at least one genre label for each entry in the dataset, we decided to manually reclassify each novel entry. We were able to conduct this classification by utilizing the data’s existing genre labels, the plot summary, as well as external research. We decided to only focus on the following genres: fiction, science fiction, fantasy, romance, historical and mystery. In doing so, it is possible that we removed some nuance and very specific genres that may describe a novel. For example, the genre “Roman a clef” is a very particular genre that appeared only about three times in the dataset, so all the information that comes with such a genre was lost. Also, there are many novels that were labeled with multiple genres, so it was up to our discretion to find the genre that fit it best out of the six we were most interested in analyzing based on the plot summary. This was particularly difficult in cases where novels had more than one genre from the main six genres we were studying. This also removed some nuance from the data as books are very complex and, thus, it is certainly within the realm of possibility that a novel may have more than one genre that describes it. We completely omitted the overlap between genres for any book in order to remove any ambiguity with the predictions.

Furthermore, to construct our persona, we needed to collect the action verbs, patient words, and the adjectives associated with the persona. A data collection choice that we made is that we decided to focus only on the main character of the plot summary. As the main character, we believed that these characters would have a more fleshed-out persona for analysis. We determined who the main character was by identifying the character with the highest frequency within the summary. This greatly limited the amount of data in our dataset. As a result, in some cases, some fields had one word, and others had none. We were able to collect and sort the action verbs, patient verbs, and adjectives through spaCy by utilizing each token’s dependencies and part of speech tags.

Limitations of our dataset is that the plot summaries are sourced from Wikipedia, meaning that they are user-generated, which could influence the accuracy of the  summaries. For example, the biases of the users may consciously or unconsciously enter the plot summaries. However, we hope that the contribution of multiple users would diminish the inaccuracies of the summaries.

As previously mentioned, we are trying to train two models to predict the genre of a novel and then compare the genre predictions produced by the two models. To accomplish this, we first developed topic models in order to understand how words from the plot summaries can be grouped together, which may provide some initial insight regarding how words from the broad plot summaries relate to genre using LDA. In order to evaluate how well the topic models can classify genres, we looked at the average cross-validated weighted F1 scores. For the persona model, we only input the action verbs, patient verbs, and the adjectives into the model to be used to classify the text into one of the six genres mentioned earlier. On the other hand, the plot model utilizes the entire plot summary as the input. Then, we performed a similar analysis using a distilBERT model, which used the same set of inputs for each respective model. Using the BERT model, we were interested in seeing whether it could classify the input text with the true labels. In order to evaluate the BERT models, we used a multiclass Logistic Regression as our baseline model and also computed performance metrics. Topic models are adept at uncovering broad topics across different texts and extracting features, whereas BERT models are proficient in capturing the nuanced meanings in text. Due to these strengths, we felt these two models were the best suited for classifying into genres.

```python
# Construct dataframe from corpus
fp = 'booksummaries.txt'
data = []

# Parse corpus to get key points of information
with open(fp, 'r', encoding = 'latin-1') as f:
  for line in f:
    val = line.strip().split('\t')
    data.append(val)
df = pd.DataFrame(data)

# Clean dataframe by renaming columns and dropping rows with any columns that don't have data
df.rename(columns={0: "id", 2: "title", 3: "author", 4: 'date_published', 5: "genres", 6: "summary"}, inplace = True)
df.drop(columns=[1], inplace = True)
df['genres'].replace('', np.nan, inplace=True)
df['date_published'].replace('', np.nan, inplace=True)
df['author'].replace('', np.nan, inplace=True)
df.dropna(subset=['genres'], inplace=True)
df.dropna(subset=['date_published'], inplace=True)
df.dropna(subset=['author'], inplace=True)

# Convert dictionaries with genres to a cleaner genre list for each book
genres = []
for i in df['genres']:
  genres.append(list(json.loads(i).values()))
df['genre_new'] = genres
df.drop(columns=['genres'], inplace = True)

# Only use texts that have the year in the date_published column
df = df[df['date_published'].str.len() == 4]

# To condense dataset further, take a random sample of 10% of the remaining data
df = df.sample(frac = 0.5, replace = False)
df
```

```python
# Load spacy model
nlp = spacy.load('en_core_web_sm')
```

```python
# Create a function that returns the most frequently mentioned character in the summary using Named Entity Recognition
def main_character(text):
  doc = nlp(text)
  persons = []
  for ent in doc.ents:
    if ent.label_ == 'PERSON':
      persons.append(ent.text)
  main = Counter(persons).most_common(1)
  return main[0][0] if main else None

# Add the main character to the dataframe as a new column
main_char = []
for i in df['summary']:
  character = main_character(i)
  main_char.append(character)
df['main_char'] = main_char
df
```

```python
# Drop rows with main_char = None
df['main_char'].replace('', np.nan, inplace=True)
df.dropna(subset=['main_char'], inplace=True)
df
```

```python
# Create a function that returns the verbs associated with what the main character is doing
def action_verbs(text, main_character):
  doc = nlp(text)
  action_verbs = []
  for chunk in doc.noun_chunks:
    if str(chunk) == main_character:
      action_verbs.append(chunk.root.head.text)
  return action_verbs

full_action_verbs = []
for i in np.arange(0, len(df)):
  list_action_verbs = action_verbs(df['summary'].iloc[i], df['main_char'].iloc[i])
  full_action_verbs.append(list_action_verbs)
df['action_verbs'] = full_action_verbs
df
```

```python
# Utilize an anonymous function to remove any action_verb lists that are empty
df = df[df['action_verbs'].apply(lambda x : len(x) > 0)]
df
```

```python
# Create a function that determines the adjectives linked with the main character
def adjectives(text, main_character):
  doc = nlp(text)
  adjs = []
  for sentence in doc.sents:
    for ent in sentence.ents:
      if ent.label_ == 'PERSON' and ent.text.lower() == main_character.lower():
        for word in sentence:
          if word.head.text == ent.text and word.pos_ == 'ADJ':
            adjs.append(word.text)
  return adjs

full_adjs = []
for i in np.arange(0, len(df)):
  list_adjs = adjectives(df['summary'].iloc[i], df['main_char'].iloc[i])
  full_adjs.append(list_adjs)
df['adjectives'] = full_adjs
df
```

```python
# Utilize an anonymous function to remove any adjectives lists that are empty
df = df[df['adjectives'].apply(lambda x : len(x) > 0)]
df
```

```python
# Create a function that returns the verbs describing actions happening to the main character
def patient_verbs(text, main_character):
  doc = nlp(text)
  patient_verbs = []
  for chunk in doc.noun_chunks:
    if str(chunk) == main_character and (chunk.root.dep_ == "dobj" or chunk.root.dep_ == "iobj" or chunk.root.dep_ == "nsubjpass"):
      patient_verbs.append(chunk.root.head.text)
  return patient_verbs

full_patient_verbs = []
for i in np.arange(0, len(df)):
  list_patient_verbs = patient_verbs(df['summary'].iloc[i], df['main_char'].iloc[i])
  full_patient_verbs.append(list_patient_verbs)

df['patient_verbs'] = full_patient_verbs
df
```

### Predict genre based on plot


```python
# Create a vectorizer for the book summary data
vectorizer = CountVectorizer(
    input = 'content',
    strip_accents = 'unicode',
    stop_words="english",
    lowercase = True,
    encoding = 'latin-1')
X = vectorizer.fit_transform(df['summary'])
```

```python
# Fit an LDA topic model on the vectorized data
lda = LatentDirichletAllocation(
    n_components = 10,
    n_jobs = -1,
    verbose = 1,
    max_iter = 5,
    evaluate_every = 0)

lda.fit(X)
```

```python
# Function pulled from Problem Set 3 (which was provided as a "freebie") to display top words from each topic
def display_topics(model, feature_names, k_top_words, title, n_topics=10):
    fig, axes = plt.subplots(round(n_topics/5), 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[-k_top_words:]
        top_features = feature_names[top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f"Topic {topic_idx +1}", fontdict={"fontsize": 30})
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in "top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Display top 5 words from each topic
display_topics(lda, vectorizer.get_feature_names_out(), k_top_words = 5, title = 'Top Words From Each Topic', n_topics = 10)
```

```python
# Because so many of the top words are character names, we decided to add main character names to the list of stopwords
main_char_list = ' '.join(df['main_char']).lower().split()
new_stopwords = list(ENGLISH_STOP_WORDS) + main_char_list
print('The new stopwords are as follows: ', new_stopwords )
```

```python
# Create a function to determine all character names in a text snippet, and then add these names to the list of stopwords
characters = []
def characters(text):
  doc = nlp(text)
  characters = []
  for ent in doc.ents:
    if ent.label_ == 'PERSON':
      characters.append(ent.text)
  return characters

for i in df['summary']:
  new_stopwords += characters(i)

new_lower_stopwords = []
for word in new_stopwords:
  new_lower_stopwords.append(word.lower())
print('The new stopwords with all character names included are as follows: ', new_lower_stopwords)
```

```python
# Reconstruct the vectorizer and run the LDA model again (pt 2)
vectorizer = CountVectorizer(
    input = 'content',
    strip_accents = 'unicode',
    stop_words = new_stopwords,
    lowercase = True,
    encoding = 'latin-1')
X = vectorizer.fit_transform(df['summary'])

lda = LatentDirichletAllocation(
    n_components = 10,
    n_jobs = -1,
    verbose = 1,
    max_iter = 5,
    evaluate_every = 0)

lda.fit(X)
```

```python
# Display the new top words per topic with the new stopwords removed and more top words displayed
display_topics(lda, vectorizer.get_feature_names_out(), k_top_words = 10, title = 'Top Words From Each Topic', n_topics = 10)
```

```python
# Generate doc-topic matrix
with warnings.catch_warnings():
  warnings.simplefilter('ignore')
  doc_topic_matrix = lda.transform(X)
print('Doc_topic_matrix shape: ', doc_topic_matrix.shape)
```

```python
# Predict genre using the 10 topic features generated above
genre = df['genre']
X_topics = StandardScaler().fit_transform(doc_topic_matrix)
predictor = LogisticRegression().fit(X_topics, genre)
y_pred = predictor.predict(X_topics)
score = np.mean(cross_val_score(predictor, X_topics, genre, scoring = 'f1_weighted'))
print('The averge cross-validated weighted f1 score is: ', score)
```

```python
# Predict genre classification using TruncatedSVD
reducer = TruncatedSVD(n_components = 10)
X_svd = reducer.fit_transform(X)
score = np.mean(cross_val_score(LogisticRegression(max_iter = 2000), X_svd, genre, scoring = 'f1_weighted'))
print('The average cross-validated weighted F1 score is: ', score)
```

```python
# Set parameters and file paths for BERT
model_name = 'distilbert-base-cased'
device_name = 'cpu'
cached_model_directory_name = 'distilbert-books'
```

```python
# Load DistilBertTokenizerFast from HuggingFace
tokenizer = DistilBertTokenizerFast.from_pretrained(model_name)
```

```python
# Encode data for BERT
# Convert labels into integers
unique_labels = set(label for label in train_genre)
labels = {label: id for id, label in enumerate(unique_labels)}
ids = {id: label for label, id in labels.items()}
print('The labels of ',labels.keys(), 'have the corresponding keys: ', ids.keys())

# Encode poems and labels
train_encodings = tokenizer(train_summary.tolist(), truncation=True, padding=True)
test_encodings  = tokenizer(test_summary.tolist(), truncation=True, padding=True)
train_genres_encoded = [labels[y] for y in train_genre]
test_genres_encoded  = [labels[y] for y in test_genre]

# Create a Torch MyDataset class
class MyDataset(torch.utils.data.Dataset):
  def __init__(self, encodings, labels):
      self.encodings = encodings
      self.labels = labels

  def __getitem__(self, idx):
      item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
      item['labels'] = torch.tensor(self.labels[idx])
      return item

  def __len__(self):
      return len(self.labels)

# Create two new dataset objects, train_dataset and test_dataset
train_dataset = MyDataset(train_encodings, train_genres_encoded)
test_dataset = MyDataset(test_encodings, test_genres_encoded)
```

```python
# Utilize compute_metrics function from Problem Set 5 to output performance metrics
def compute_metrics(pred):
  labels = pred.label_ids
  preds = pred.predictions.argmax(-1)
  precision = precision_score(labels, preds, average = 'weighted')
  recall = recall_score(labels, preds, average = 'weighted')
  accuracy = accuracy_score(labels, preds)
  f1 = f1_score(labels, preds, average='weighted')
  return {
      'precision': precision,
      'recall': recall,
      'accuracy': accuracy,
      'f1': f1
  }

# Create trainer object
trainer = Trainer(
  model = model,
  args = training_args,
  train_dataset = train_dataset,
  eval_dataset = test_dataset,
  compute_metrics = compute_metrics
)
```

## 3. Results

When using plot summaries to predict genre labels, through the top 10 topic features, we were able to achieve an average cross-validated weighted f1 score of about 0.12088. On the other hand, we were able to achieve higher scores when using a TruncatedSVD approach, in which the average cross-validated weighted f1 score was about 0.32958. We believe the higher TruncatedSVD score is likely due to the significant amount of noise in the data, which is minimized by using the TruncatedSVD approach that lowers the dimensionality of the data. The impact of this noise is exacerbated by the smaller size of the dataset, so the amount of variation in the data cannot be smoothed as done in larger datasets. When using the BERT classification method, our trainer.train() line was able to display how our training loss and f1 scores changed as the model was trained (please see output above). Clearly, there is a steady decline in the training loss and a steady rise in the f1 scores. We were able to achieve an accuracy score of 0.48, which is greater than the baseline score of 0.39, and an average weighted score of 0.35, which is approximately the same as the score produced via our baseline logistic regression model. However, again, we believe that the scores had significant variability due to the small size of the dataset, and we anticipate stronger results if using the original dataset. In terms of which genres generated the best f1 scores, we found that fiction had a f1 score of 0.63 and scifi had a f1 score of 0.59, likely due to the presence of more texts with these genre labels in the testing set. Finally, we were able to display a heat map to demonstrate the differences between the true and predicted genre for the fiction and scifi texts. Unsurprisingly, texts from a multitude of genres were incorrectly classified as fiction, which is likely because fiction and other genres have a lot of overlap. For scifi texts, the true genre was fantasy, which makes sense considering that scifi and fantasy novels have similar usage of fictional vocabulary.

When using personas to predict genre labels, we were able to achieve better results through the top 10 topic features as compared to the plot summary results, with an average cross-validated weighted f1 score of about 0.20213. This also outperforms its TruncatedSVD counterpart, which generated a score of about 0.18815. As for the BERT analysis, we were again able to see a fall in the validation loss and a rise in the f1 score while the model was training. The average score of 0.52 is significantly greater than the baseline average score of 0.21, and the weighted average score of 0.41 is also much higher than the baseline weighted average score of 0.20. This suggests improved classification performance via the BERT model, and while fiction and scifi again achieved high f1 scores (0.63 and 0.62, respectively), fantasy was also able to achieve a high score of 0.50. With fiction and scifi, we see similar misclassifications as with the plot summaries through the heat map, again likely due to these genres’ overlap with adjacent genres.

One limitation of our study is that the corpus we assembled to use with our two models was rather small; most of the data from the original dataset was not used in our study.
